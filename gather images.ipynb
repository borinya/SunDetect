```python
df = pd.read_csv('./sun_disk_pos_database.2021-03-06.csv', sep=';')
```

```python
all_images_dict = dict([(os.path.basename(s), s) for s in all_images_list])
```

```python
missed_fnames = []
```

```python
def full_fname(basename):
    try:
        return all_images_dict[basename]
    except:
        missed_fnames.append(basename)
        return None
```

```python
labeled_images = [full_fname(fn) for fn in df.img_filename]
```

```python
labeled_images
```

```python
for fn in tqdm(labeled_images):
    new_fname = os.path.join('./images', os.path.basename(fn))
    shutil.copyfile(fn, new_fname)
```
